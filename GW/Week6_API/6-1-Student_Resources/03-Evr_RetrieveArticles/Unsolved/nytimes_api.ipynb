```python
# Dependencies
import requests
from config import api_key
from pprint import pprint
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
query = "coronavirus"
```

```python
# Search for articles published between a begin and end date
query_url = url + "api-key=" + api_key + "&q="+query+"&begin_date="+"20200101"+"&end_date="+"20211123"

# Build query url
data = requests.get(query_url).json()
data["response"]["docs"]
```

```python
for i in range(10):
    print(data["response"]["docs"][i]["snippet"])
    print("\n")
    
```

```python
new_query = query_url + "&page="+"30"
new_data = requests.get(new_query).json()
for i in range(10):
    
    print(new_data["response"]["docs"][i]["snippet"])
    print("---")
```
