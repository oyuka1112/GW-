```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
type(cities)
```

```python
# Create a bar chart based upon the above data
plt.figure(dpi = 40)
plt.bar( x_axis, bars_in_cities, alpha = 0.5, color = 'blue')
plt.xticks(x_axis, cities)
# plt.xlim(0, 4)
# plt.ylim(0, 10)
plt.xlabel("Cities")
plt.ylabel(" Bar ")
plt.title("Number of bars in the city")

# plt.savefig("png")
```
