```python
y = [10, 30, 50, 45, 24]
x= np.arange(len(y))
plt.figure(figsize = (5, 4))
plt.bar(x, y, color = "blue", alpha= 0.5)
plt.xticks(x, ['Java', "C", "P", "R", "Cl"]) #name x
plt.xlim(0)
```

```python
#Pie Chart
labels = ["Humans", "Smurf", "Hobbits", "Ninjas"]
size = [220, 95, 80, 100]
colors = ['red', 'orange', 'blue', 'lightskyblue']
explode = (0.1, 0, 0, 0)
plt.pie(size, labels= labels, colors = colors, startangle = 90, explode = explode )
```
