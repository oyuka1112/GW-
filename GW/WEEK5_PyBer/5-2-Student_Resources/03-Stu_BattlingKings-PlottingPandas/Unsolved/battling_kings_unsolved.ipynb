```python
# Read CSV
df1 = pd.read_csv("Resources/got.csv")
df1
```

```python
# Get attacker and defender king data
df = df1[["attacker_king", "defender_king"]]
df.head()
```

```python
# Get total battle data
defender_king_df = df.groupby(["defender_king"]).count()
attacker_king_df = df.groupby(["attacker_king"]).count()
defender_king_df = defender_king_df.rename(columns = {"attacker_king":"count", "defender_king":"king"})
attacker_king_df = attacker_king_df.rename(columns = {"defender_king":"count","attacker_king":"king" })
```
